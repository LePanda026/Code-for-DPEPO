```python
import uuid
import numpy as np

batch_size = 8
n = 4

group_ids = []
uid_batch = []
for i in range(batch_size):
    if i % n == 0:
        group_id = 0
        uid = str(uuid.uuid4())
    else:
        group_id += 1
    group_ids.append(group_id)
    uid_batch.append(uid)
uid_batch = np.array(uid_batch, dtype=object) 
group_ids = np.array(group_ids, dtype=object) 

uid_batch,group_ids 
```
