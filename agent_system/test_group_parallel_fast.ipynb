```python
import json
from env_manager_parallel import build_parallel_alfworld_envs 

data = json.load(open('/data/home/zhangjs/disk/project/verl-agent/data_pipelines/gamefiles/gamefiles_train.json','r'))

gamefiles = list(data.values()) 
```

```python
import time
from tqdm import tqdm
step = 16

for i in tqdm(range(0,3000,step)):
    start_time = time.time()
    parallel_env = build_parallel_alfworld_envs(gamefiles=gamefiles[i:i+step],
                                                # env_num=512,
                                                group_n=8,
                                                resources_per_worker={'num_cpus': 0.5},
                                                num_parallel=5) 
    parallel_env.reset() 
    print(time.time() - start_time) 
```
