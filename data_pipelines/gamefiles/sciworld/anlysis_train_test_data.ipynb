```python
import json 
import random
from collections import defaultdict

def read_json(file_path):
    data = json.load(open(file_path,'r'))
    return data 

def write_json(data, path):
    with open(path, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
```

```python
import random 
sample_ratio = 1 
min_samples = 1 
random.seed(42) 


l2_data = read_json('L2_idx.json') 
l2_train_data = l2_data['train'] 

# 按第一个元素分组 
groups = defaultdict(list)
for item in l2_train_data:
    groups[item[0]].append(item)

sampled = []

for task, items in groups.items(): 
    n = max(min_samples, int(len(items) * sample_ratio)) 
    n = min(n, len(items)) 
    sampled.extend(random.sample(items, n)) 

from scienceworld import ScienceWorldEnv 

env = ScienceWorldEnv("", "", envStepLimit=50)
task_names = env.get_task_names()

sampled_list = [] 
for sample in sampled: 
    task_id = sample[0]
    var_id  = sample[1]

    task_name = task_names[task_id]
    sampled_list.append({'task':task_name, 'id': var_id})

sampled_list[0] 

write_json(sampled_list,'/data/home/zhangjs/disk/project/verl-agent/data_pipelines/gamefiles/sciworld/test/L2_full.json') 
```

```python
import random 
from scienceworld import ScienceWorldEnv 

sample_ratio = 1
min_samples = 1 
random.seed(42) 

l0_data = read_json('L0_idx.json') 
l0_train_data = l0_data['train'] 

l2_train_data = read_json('L2_idx.json')['train']

dedu_data = set([tuple(elem) for elem in l0_train_data]) - set([tuple(elem) for elem in l2_train_data])
dedu_data = [list(elem) for elem in dedu_data]

print(len(dedu_data))

groups = defaultdict(list)
for item in dedu_data:
    groups[item[0]].append(item)

L01_data = []
for task, items in groups.items(): 
    n = max(min_samples, int(len(items) * sample_ratio)) 
    n = min(n, len(items)) 
    L01_data.extend(random.sample(items, n)) 

env = ScienceWorldEnv("", "", envStepLimit=50)
task_names = env.get_task_names()
 
sampled_list = []
for sample in L01_data:
    task_id = sample[0]
    var_id  = sample[1]

    task_name = task_names[task_id]
    sampled_list.append({'task':task_name, 'id': var_id})

sampled_list[0] 

write_json(sampled_list,'/data/home/zhangjs/disk/project/verl-agent/data_pipelines/gamefiles/sciworld/test/L_dedu_1.json') 
```
